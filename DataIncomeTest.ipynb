```python
import numpy as np
import pandas as pd
```

```python
datawith = pd.read_csv('with.csv')
datawith.loc[datawith['Married']=="Yes",'Married']=1
datawith.loc[datawith['Married']=='No','Married']=0
datawith.loc[datawith['Education']=="Graduate",'Education']=1
datawith.loc[datawith['Education']=='Not Graduate','Education']=0
datawith.loc[datawith['Self_Employed']=="Yes",'Self_Employed']=1
datawith.loc[datawith['Self_Employed']=='No','Self_Employed']=0
datawith.loc[datawith['Property_Area']=="Urban",'Property_Area']=0
datawith.loc[datawith['Property_Area']=='Semiurban','Property_Area']=1
datawith.loc[datawith['Property_Area']=='Rural', 'Property_Area']=2
datawith.head()
```

```python
datawo = pd.read_csv('without.csv')
datawo.loc[datawo['Married']=="Yes",'Married']=1
datawo.loc[datawo['Married']=='No','Married']=0
datawo.loc[datawo['Education']=="Graduate",'Education']=1
datawo.loc[datawo['Education']=='Not Graduate','Education']=0
datawo.loc[datawo['Self_Employed']=="Yes",'Self_Employed']=1
datawo.loc[datawo['Self_Employed']=='No','Self_Employed']=0
datawo.loc[datawo['Property_Area']=="Urban",'Property_Area']=0
datawo.loc[datawo['Property_Area']=='Semiurban','Property_Area']=1
datawo.loc[datawo['Property_Area']=='Rural', 'Property_Area']=2
datawo.head()
```

```python
X1 = datawith.drop("Loan_Status", axis=1)
y1 = datawith["Loan_Status"]
print(X1.shape, y1.shape)
```

```python
X2 = datawo.drop("Loan_Status", axis=1)
y2 = datawo["Loan_Status"]
print(X2.shape, y2.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, random_state=1, stratify=y1)
X1_scaler = StandardScaler().fit(X1_train)
X1_train_scaled = X1_scaler.transform(X1_train)
X1_test_scaled = X1_scaler.transform(X1_test)


# Step 1: Label-encode data set
label1_encoder = LabelEncoder()
label1_encoder.fit(y1_train)
encoded_y1_train = label1_encoder.transform(y1_train)
encoded_y1_test = label1_encoder.transform(y1_test)

# Step 2: Convert encoded labels to one-hot-encoding
y1_train_categorical = to_categorical(encoded_y1_train)
y1_test_categorical = to_categorical(encoded_y1_test)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, random_state=1, stratify=y2)
X2_scaler = StandardScaler().fit(X2_train)
X2_train_scaled = X2_scaler.transform(X2_train)
X2_test_scaled = X2_scaler.transform(X2_test)


# Step 1: Label-encode data set
label2_encoder = LabelEncoder()
label2_encoder.fit(y2_train)
encoded_y2_train = label2_encoder.transform(y2_train)
encoded_y2_test = label2_encoder.transform(y2_test)

# Step 2: Convert encoded labels to one-hot-encoding
y2_train_categorical = to_categorical(encoded_y2_train)
y2_test_categorical = to_categorical(encoded_y2_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model1 = Sequential()
model1.add(Dense(units=100, activation='relu', input_dim=11))
model1.add(Dense(units=100, activation='relu'))
model1.add(Dense(units=2, activation='softmax'))

model2 = Sequential()
model2.add(Dense(units=100, activation='relu', input_dim=11))
model2.add(Dense(units=100, activation='relu'))
model2.add(Dense(units=2, activation='softmax'))
```

```python
model1.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
model1.fit(
    X1_train_scaled,
    y1_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model2.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
model2.fit(
    X2_train_scaled,
    y2_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model1_loss, model1_accuracy = model1.evaluate(X1_test_scaled, y1_test_categorical, verbose=2)
print(f"Normal Neural Network with data - Loss: {model1_loss}, Accuracy: {model1_accuracy}")

model2_loss, model2_accuracy = model2.evaluate(X2_test_scaled, y2_test_categorical, verbose=2)
print(f"Normal Neural Network with/out data - Loss: {model2_loss}, Accuracy: {model2_accuracy}")
```
